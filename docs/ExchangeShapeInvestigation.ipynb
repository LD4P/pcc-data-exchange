## Directly from Sinopia

```python
work_result = requests.get("https://api.stage.sinopia.io/resource/181d118d-cce3-4506-811c-2e4a1eaf50b8")
instance_result = requests.get("https://api.stage.sinopia.io/resource/2559d178-6c72-47d1-9042-c547e1b4e0e6")
```

```python
sinopia_graph = rdflib.Graph()
for prefix, url in [
    ("bf","http://id.loc.gov/ontologies/bibframe/"),
    ("bflc", "http://id.loc.gov/ontologies/bflc/"),
    ("rdfs", "http://www.w3.org/2000/01/rdf-schema#"),
    ("sinopia","http://sinopia.io/vocabulary/")
]:
    sinopia_graph.namespace_manager.bind(prefix,url)
```

```python
sinopia_graph.parse(data=json.dumps(work_result.json().get("data")), format='json-ld')
sinopia_graph.parse(data=json.dumps(instance_result.json().get("data")), format='json-ld')
```
